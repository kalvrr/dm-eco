```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 0.33  # Cobb-Douglas production exponent
n = 0.02      # Population growth rate
g = 0.02      # Technological progress rate
delta = 0.05  # Depreciation rate
s_gr = alpha  # Golden rule savings rate (approximately alpha for Cobb-Douglas)



# Functions
def production_function(k):
    return k ** alpha

def savings_function(k, s):
    return s * production_function(k)

def depreciation_function(k):
    return (n + g + delta) * k

# Capital grid
k = np.linspace(0.01, 10, 500)

# Calculate functions
output = production_function(k)
savings_gr = savings_function(k, s_gr)
depreciation = depreciation_function(k)

# Golden rule capital stock
k_gr = (alpha / (n + g + delta)) ** (1 / (1 - alpha))
output_gr = production_function(k_gr)
savings_at_gr = savings_function(k_gr, s_gr)
consumption_at_gr = output_gr - savings_at_gr
depreciation_gr = depreciation_function(k_gr)

def tangent_function(k,output_gr,depreciation_gr):
    return (output_gr-depreciation_gr)+(n + g + delta) * k

tangent = tangent_function(k,output_gr,depreciation_gr)

# Plot
plt.figure(figsize=(10, 6))

plt.annotate("", xy=(10.2, 0), xytext=(0, 0),
                 arrowprops=dict(arrowstyle="->", color='black', lw=1.5))

plt.annotate("", xy=(0, 2.2), xytext=(0, 0),
                 arrowprops=dict(arrowstyle="->", color='black', lw=1.5))
plt.text(4.1, -0.075, "$\\kappa$", fontsize=12, color='black')  # Label for x-axis arrow
plt.text(-0.3, 0.3, "$f(\\kappa)$, $sf(\\kappa)$ and $(n+g+\\delta)\\kappa$", fontsize=12, color='black', rotation=90)  # Label for y-axis arrow

# Curves
plt.plot(k, output, label="Output = $f(\\kappa)$", color="purple")
plt.plot(k, savings_gr, label="Savings at the golden rule = ${s_{gr} f(\\kappa_t)}$", color="blue")
plt.plot(k, depreciation, label="Dilution plus depreciation = $(n+g+\\delta)\\kappa_t$", color="black", linestyle="--")
plt.plot(k, tangent, label="Tangent w/ slope $(n+g+\\delta)$", color="grey", linestyle="--")
#plt.plot([k_gr, k_gr], [savings_at_gr, output_gr], color="red", linestyle="-", linewidth=2)

# Highlight golden rule
plt.axvline(x=k_gr, color="gray", linestyle="--")
plt.scatter([k_gr], [output_gr], color="red")
plt.scatter([k_gr], [savings_at_gr], color="blue")
#plt.scatter([k_gr], [consumption_at_gr], color="red")

# Annotate golden rule points
plt.text(k_gr + 0.2, output_gr-0.1, "A", fontsize=12,color='purple')
plt.text(k_gr + 0.2, savings_at_gr+0.1, "B", fontsize=12,color='blue')
#plt.text(k_gr + 0.2, consumption_at_gr-0.1, "C", fontsize=12)

# Labels and annotations
#plt.title("Golden Rule of Capital in the Solow Model", fontsize=16)
plt.xlabel("Capital per effective worker", fontsize=14)
plt.ylabel("", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend(fontsize=12)

# Annotations for arrows and labels
plt.annotate("", xy=(k_gr, depreciation_function(k_gr)), xytext=(k_gr, production_function(k_gr)),
                 arrowprops=dict(arrowstyle="<->", color='red', lw=1.5))

plt.annotate("Steady-state consumption\nat the golden rule\n= $(1-s_{gr})f(\\kappa^*_{gr})$",
             xy=(k_gr, consumption_at_gr), xytext=(k_gr + 0.5, consumption_at_gr),
             arrowprops=dict(facecolor='black', arrowstyle="->"), fontsize=12,color='red')

plt.annotate("Steady-state savings at\nthe golden rule = $s_{gr} f(\\kappa^*_{gr})$",
             xy=(k_gr, savings_at_gr), xytext=(k_gr +0.5, savings_at_gr - 0.5),
             arrowprops=dict(facecolor='blue', arrowstyle="->"), fontsize=12)

plt.annotate("Straight line of\nslope n+g+\u03b4",
             xy=(4, depreciation_function(4)), xytext=(2, 1),
             arrowprops=dict(facecolor='black', arrowstyle="->"), fontsize=12)

plt.annotate("Savings of\nslope n+g+\u03b4",
             xy=(2, savings_function(2,s_gr)), xytext=(4, 0.15),
             arrowprops=dict(facecolor='black', arrowstyle="->"), fontsize=12,color='blue')

plt.annotate("$\\varphi=f(\\kappa)$",
             xy=(6, production_function(6)), xytext=(5, 2.1),
             arrowprops=dict(facecolor='black', arrowstyle="->"), fontsize=12, color='purple')

plt.text(k_gr+0.2, -0.075, "$\\kappa^*_{gr}$", fontsize=12, ha='center',color='grey')

#plt.annotate("Steady-state consumption\nat the golden rule\n= $(1-s_{gr})f(\\kappa^*_gr)$",
#            xy=(0.7, 0.1), xycoords='figure fraction',
#            xytext=(0.7, 0.5), textcoords='figure fraction',
#            arrowprops=dict(arrowstyle="-",lw=0), fontsize=12)
# Display
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("golden_rule.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 0.33  # Cobb-Douglas production exponent
n = 0.02      # Population growth rate
g = 0.02      # Technological progress rate
delta = 0.05  # Depreciation rate
s_gr = alpha  # Golden rule savings rate (approximately alpha for Cobb-Douglas)

# Functions
def production_function(k):
    return k ** alpha

def savings_function(k, s):
    return s * production_function(k)

def depreciation_function(k):
    return (n + g + delta) * k

# Capital grid
k = np.linspace(0.01, 10, 500)

# Calculate functions
output = production_function(k)
savings_gr = savings_function(k, s_gr)
depreciation = depreciation_function(k)

# Golden rule capital stock
k_gr = (alpha / (n + g + delta)) ** (1 / (1 - alpha))
output_at_gr = production_function(k_gr)
savings_at_gr = savings_function(k_gr, s_gr)
consumption_at_gr = output_at_gr - savings_at_gr

# Plot
plt.figure(figsize=(10, 8))

# Curves
plt.plot(k, output, label="Output = f(k_t)", color="black")
plt.plot(k, savings_gr, label="Savings at the golden rule = s_gr f(k_t)", color="blue")
plt.plot(k, depreciation, label="Dilution plus depreciation = (n+g+\u03b4)k_t", color="black", linestyle="--")
plt.plot([k_gr, k_gr], [savings_at_gr, output_at_gr], color="red", linestyle="-", linewidth=1)

# Highlight golden rule
plt.axvline(x=k_gr, color="gray", linestyle="--")
plt.scatter([k_gr], [output_at_gr], color="black")
plt.scatter([k_gr], [savings_at_gr], color="blue")
plt.scatter([k_gr], [consumption_at_gr], color="red")

# Annotate golden rule points
plt.text(k_gr + 0.2, output_gr, "A", fontsize=12)
plt.text(k_gr + 0.2, savings_at_gr, "B", fontsize=12)
plt.text(k_gr + 0.2, consumption_at_gr, "C", fontsize=12)

# Add bracket between A and B
plt.plot([k_gr, k_gr], [savings_at_gr, output_at_gr], color="black", linestyle="-", linewidth=1)
plt.text(k_gr + 0.3, (savings_at_gr + output_at_gr) / 2, "Steady-state\nconsumption", fontsize=12, va='center')

# Labels and annotations
plt.title("Golden Rule of Capital in the Solow Model", fontsize=16)
plt.xlabel("k_t", fontsize=14)
plt.ylabel("", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend(fontsize=12)

# Annotations for arrows and labels
#plt.annotate("Steady-state consumption\nat the golden rule\n= (1-s_gr)f(k*_gr)",
#             xy=(k_gr, consumption_at_gr), xytext=(k_gr + 1, consumption_at_gr),
#             arrowprops=dict(facecolor='black', arrowstyle="->"), fontsize=12)

plt.annotate("Steady-state consumption\nat the golden rule\n= (1-s_gr)f(k*_gr)",
            xy=(k_gr, consumption_at_gr), xycoords='figure fraction',
            xytext=(k_gr, output_at_gr), textcoords='figure fraction',
            arrowprops=dict(arrowstyle="-",lw=0
                       ))

plt.annotate("Steady-state savings at\nthe golden rule = $s_gr f(\\kappa^*_{gr})$",
             xy=(k_gr, savings_at_gr), xytext=(k_gr +0.5, savings_at_gr - 0.5),
             arrowprops=dict(facecolor='blue', arrowstyle="->"), fontsize=12)

plt.annotate("Straight line of slope n+g+\u03b4",
             xy=(5, depreciation_function(5)), xytext=(3, 1),
             arrowprops=dict(facecolor='black', arrowstyle="->"), fontsize=12)

plt.annotate("Output = f(k_t)",
             xy=(6, production_function(6)), xytext=(5, 2.1),
             arrowprops=dict(facecolor='black', arrowstyle="->"), fontsize=12)

plt.text(k_gr, -0.5, "Steady-state capital stock\nat the golden rule = $\\kappa^*_{gr}$", fontsize=12, ha='center')

# Display
plt.grid(alpha=0.3)
#plt.tight_layout()
plt.show()
```
